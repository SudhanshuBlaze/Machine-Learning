## Decision Trees works best when --> both independent variables and target variable are Classified values

**Logistic regression used when independent variables have continuous values and dependent(target) variables have classified values**

**Linear regression used when both independent variables and dependent(target) variables have classified value**

```python
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
data.data
```

```python
data.feature_names   #prints feature names
```

```python
data.target  #prints target Values
```

```python
data.target_names   #prints target Names
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names)+['target']])
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
X=df.iloc[:,0:-1]   #this does not contains target
y=df.iloc[:,-1]     #this contains target only
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print('Shape of X_train ',X_train.shape)
print('Shape of y_train ',y_train.shape)
print('Shape of X_test ',X_test.shape)
print('Shape of y_test ',y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini')  #gini is used by default
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model1=DecisionTreeClassifier(criterion='entropy')
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

## Steps to create a Random Forest

### Step 1: Create a Bootstrapped Dataset




**Bootstrapped dataset:** randomly picking samples from the dataset with **replacement**. As you can see in the image we have picked data from row 4 twice, so this is what meant by **replacement**.

Benefit: Decisions trees are very sensitive to the data they are trained on — small changes to the training set can result in significantly different tree structures. Random forest takes advantage of this by allowing each individual tree to randomly sample from the dataset with replacement, **resulting in different trees**.

### Bagging (Bootstrap Aggregation) :
Bootstrapping the data + using the aggregate(vote) to make decision is known as Bagging

**Aim:** to achieve less variance than any model has individually.

### Step 2: 
Feature Randomness — In a normal decision tree, when it is time to split a node, we consider every possible feature(or columns) which means we choose from the full list of features, even if we have already used some of those features and pick the one that produces the most separation between the observations in the left node vs. those in the right node. Thus, a single feature can appear multiple times in a tree.


- create a decision tree but only use random subset of variables(or columns) at each step for node splitting decision
- However, a random forest can pick only from a random subset of features to split a node at each step
- However, in a Random forest a single feature cannot appear multiple times in a tree.

**Reason:** This forces even more variation amongst the trees in the model and ultimately results in **lower correlation across trees and more diversification.**






**And we just build tree as usual, but only considering a random subset of variables(or columns) at each step.**

### Step 3: Repeat the steps 1 and 2 for n-times, where n is the number of decision tree you want

### Conclusion:

- The trees of the forest and more importantly their predictions need to be **uncorrelated** (or at least have low correlations with each other). 
- While the algorithm itself via **feature randomness tries to engineer these low correlations** for us, the features we select and the hyper-parameters we choose will impact the ultimate correlations as well.
