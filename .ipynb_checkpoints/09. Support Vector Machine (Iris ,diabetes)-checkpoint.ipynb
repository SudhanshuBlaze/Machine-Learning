### WORKING
At first approximation what SVMs do is to find a separating line(or hyperplane) between data of two classes. SVM is an algorithm that takes the data as an input and outputs a line that separates those classes if possible.


According to the SVM algorithm we find the points closest to the line from both the classes.These points are called support vectors. Now, we compute the distance between the line and the support vectors. This distance is called the margin. **Our goal is to maximize the margin.** The hyperplane for which the margin is maximum is the optimal hyperplane.

Thus for a space of n dimensions we have a hyperplane of **n-1 dimensions** separating it into two parts


## SVM Kernel Functions

**SVM algorithms use a group of mathematical functions that are known as kernels. The function of a kernel is to require data as input and transform it into the desired form.**

Different SVM algorithms use differing kinds of kernel functions. These functions are of different kindsfor instance, linear, nonlinear, polynomial, radial basis function (RBF), and sigmoid.

The most preferred kind of kernel function is RBF. Because it's localized and has a finite response along the complete x-axis.

## Tuning Hyper Params:
### C
It controls the trade off between smooth decision boundary and classifying training points correctly. A large value of c means you will get more training points correctly. **Higher the value of C -> Causes more accurate fit -> might Overfit**

### gamma
If gamma has a very high value, then the decision boundary is just going to be dependent upon the points that are very close to the line which effectively results in ignoring some of the points that are very far from the decision boundary. This is because the closer points get more weight and it results in a wiggly curve as shown in previous graph.On the other hand, if the gamma value is low even the far away points get considerable weight and we get a more linear curve.

```python
X=df.drop('Outcome',axis=1)
X.head()
```

```python
y=df.Outcome
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
len(X_train)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,y_train)   #training the Model
```

```python
model.score(X_test,y_test)
```

```python
model=SVC(gamma=10.0)     #gamma
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=SVC(C=2)    #regularization
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=SVC(C=5.0)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
 #Kernel -->{'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'}

model=SVC(kernel='linear')  
model.fit(X_train,y_train) 
model.score(X_test,y_test)
```

## Iris Data set

```python
import sklearn.datasets
```

```python
data=sklearn.datasets.load_iris()
```

```python
dir(data)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df["outcome"]=data.target
```

```python
data.target_names   #returns an array of flower(Target) names
```

```python
data.target_names[0]   #grabbing a value from array
```

### adding a column named flower and then putting the flower names accordingly from th outcome values



```python
df["flower"]=data.target_names[df.outcome]
```

```python
df[df["outcome"]==1]     #prints the value for all outcome=1
```

```python
#removing the outcome column as we now have flower(Target) column
df.drop("outcome",axis=1,inplace=True)
```

```python
df.head()   #final
```

```python
X=df.drop("flower",axis=1)
y=df.flower
```

### Visualisation

```python
import matplotlib.pyplot as plt
```

```python
df0=df[df["flower"]=="setosa"]
df1=df[df["flower"]=="virginica"]
```

```python
plt.scatter(df0["sepal length (cm)"],df0["sepal width (cm)"],color="red",marker="+")
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],color="green",marker="*")

plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()   #different parameters can be passed here
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test) 
```

### Tuning and tweaking the model

```python
#linear kernel is used as there is 
model=SVC(kernel="linear")  #tweaking the model 
model.fit(X_train,y_train)
model.score(X_test,y_test) 
```

```python
model=SVC(C=10)   #tweaking the model 
model.fit(X_train,y_train)
model.score(X_test,y_test) 
```

```python
#in most cases on increasing the gamma value --> score decreases
model=SVC(gamma=1)   #tweaking the model 
model.fit(X_train,y_train)
model.score(X_test,y_test) 
```
